## Списки - хранилище для произвольного числа элементов

В следующих заданиях Вам понадобятся операции со списками:
* Список задается квадратными скобками: letters = ["a", "b", "c"]. Внутри элементы перечисляются через запятую.
* Достать элемент из списка: letters[1] вернет "a". Нумерация идет с 0.
* Соединить все строки в списке через пробел: " ".join(letters) вернет "a b c".  Вместо пробела можно использовать любую строку.
* Добавить элемент в список: letters.append("d")

```python
letters = ["a", "b", "c"]
print(letters[1])
print(" ".join(letters))
```

## Циклы - повторение похожих действий

В следующих задачах Вам понадобятся циклы:
* Цикл задается с помощью команды for, например: for number in [1, 2, 3]:
* Все, что выполняется внутри цикла, пишется с отступом в 4 пробела (или Tab)
* Цикл позволяет перебирать элементы последовательности (например, списка) и повторять для них одни и те же действия.
Пример:

```python
for number in [1, 2, 3]:
    print("Hello!")
    print(number)
```

Ниже даны решения каждого пункта как изменение базовой версии, то есть изменения не накапливаются. Желающие могут собрать все изменения в одной ячейке. Решения предполагают, что перед запуском ячейки вызвана ячейка создания переменных persons, letter_rows и wishes_all.

```python
# пункт 1
for person in persons:
    name = person[0]
    gender = person[1]
    year = person[2]
    if gender == "ж":
        intro = "Дорогая " + name + "!\n"
    else:
        intro = "Дорогой " + name + "!\n"
    letter = intro + "".join(letter_rows)
    print(letter)
    print()
```

```python
# пункт 2
for person in persons:
    name = person[0]
    year = person[2]
    intro = "Добрый день, " + name + "! \n"
    intro2 = "Мы работаем с Вами с "+str(year)+" года и надеемся на дальнейшее сотрудничество.\n" 
    letter = intro + intro2 + "".join(letter_rows)
    print(letter)
    print()
```

```python
# пункт 3
for person in persons:
    name = person[0]
    year = person[2]
    intro = "Добрый день, " + name + "! \n"
    period = 2020 - year
    if period == 1:
        postfix = " год"
    elif period == 2 or period == 3 or period == 4:
        postfix = " года"
    else:
        postfix = " лет"
    intro2 = "Мы работаем с вами уже "+str(period)+postfix+" и надеемся на дальнейшее сотрудничество.\n" 
    letter = intro + intro2 + "".join(letter_rows)
    print(letter)
    print()
```

```python
# пункт 4
import numpy as np

for person in persons:
    name = person[0]
    intro = "Добрый день, " + name + "! \n"
    wishes = []
    for wish in np.random.choice(wishes_all, 3, replace=False):
        wishes.append(wish)
    wish_sent = "Желаем " + ", ".join(wishes) + "!\n"
    letter_rows[1] = wish_sent
    letter = intro + "".join(letter_rows)
    print(letter)
    print()
```
